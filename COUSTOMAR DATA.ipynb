```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("SAYANI.X.xlsx")
df=pd.DataFrame(data)
print(df.head())
grouped_by=df.groupby("channel")["cost"].sum()
print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values)
plt.xlabel("channel",fontsize=20)
plt.ylabel("cost",fontsize=20)
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("SAYANI.X.xlsx")
df=pd.DataFrame(data)
print(df.head())
grouped_by=df.groupby("channel")["revenue"].sum()
print(grouped_by)
plt.plot(grouped_by.index,grouped_by.values)
plt.xlabel("channel",fontsize=20)
plt.ylabel("cost",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("SAYANI.X.xlsx")
df=pd.DataFrame(data)
print(df.head())
plt.hist(df["conversion_rate"],bins=10)
plt.xlabel("conversion _rate",fontsize=20)
plt.ylabel("channel",fontsize=20)
plt.show()
```
